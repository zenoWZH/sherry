```python
df_new = pd.DataFrame()
df_new['blank'] = ['RF', 'SVM', 'GBDT', 'XGB']
df_new
```

```python
count = 0
for i in range(1,6):
    for j in range(i+1,10-1):
        if (j-i>5):
            break
        for k in range(j+1,10):
            if (k-j)>5:
                break
            if (10-k)>5:
                continue
            #print("%s %s %s %s" % (i, j-i, k-j, 10-k))
            count+=1
            df_new[str(count)] = pd.Series([i/10.0, (j-i)/10.0, (k-j)/10.0, (10-k)/10.0])
```
